```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

olympics_data = pd.read_csv('./Data/athlete_events.csv');

# Escludi le competizioni artistiche
df = olympics_data[olympics_data['Sport'] != 'Art Competitions']

# Ricodifica gli anni dei Giochi Olimpici invernali dopo il 1992
# per abbinarli ai successivi Giochi Olimpici estivi
original = [1994, 1998, 2002, 2006, 2010, 2014]
new = [1996, 2000, 2004, 2008, 2012, 2016]
for orig, new_year in zip(original, new):
    df['Year'] = df['Year'].replace(orig, new_year)

# Pulizia dei dati
# Rimuovere eventuali righe con valori mancanti per le colonne 'Age' e 'Sex'
df_cleaned = df.dropna(subset=['Age', 'Sex'])

# Creazione dell'istogramma per la distribuzione delle età degli atleti
plt.figure(figsize=(12, 6))
plt.hist(df_cleaned['Age'], bins=range(10, 70, 1), edgecolor='black')
plt.title('Distribuzione delle Età degli Atleti')
plt.xlabel('Età')
plt.ylabel('Numero di Atleti')
plt.grid(axis='y')
plt.show()
```

```python
# Creazione del grafico a torta per la distribuzione degli atleti per sesso
sex_counts = df_cleaned['Sex'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66b3ff', '#ff9999'])
plt.title('Distribuzione degli Atleti per Sesso')
plt.show()
```

```python
# Creazione del grafico a barre per la distribuzione degli atleti per sesso
plt.figure(figsize=(8, 6))
sex_counts.plot(kind='bar', color=['#66b3ff', '#ff9999'])
plt.title('Distribuzione degli Atleti per Sesso')
plt.xlabel('Sesso')
plt.ylabel('Numero di Atleti')
plt.xticks(rotation=0)
plt.show()
```

```python
# Numero di medaglie per evento sportivo
medals_per_event = df_cleaned['Sport'].value_counts()

# Creazione del grafico a barre per il numero di medaglie per evento sportivo
plt.figure(figsize=(12, 8))
medals_per_event.plot(kind='bar')
plt.title('Numero di Medaglie per Evento Sportivo')
plt.xlabel('Evento Sportivo')
plt.ylabel('Numero di Medaglie')
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribuzione delle medaglie (oro, argento, bronzo) per ogni evento sportivo
medals_distribution = df_cleaned.pivot_table(index='Sport', columns='Medal', aggfunc='size', fill_value=0)

# Creazione del grafico a barre per la distribuzione delle medaglie per evento sportivo
medals_distribution.plot(kind='bar', stacked=True, figsize=(12, 10), color=['#FFD700', '#C0C0C0', '#CD7F32'])
plt.title('Distribuzione delle Medaglie per Evento Sportivo')
plt.xlabel('Evento Sportivo')
plt.ylabel('Numero di Medaglie')
plt.xticks(rotation=90)
plt.legend(title='Medal')
plt.show()
```

```python
# Filter data for Italian athletes
italy_df = df[df['NOC'] == 'ITA']

# Clean data by removing rows with missing values for specific columns
italy_df_cleaned = italy_df.dropna(subset=['Age', 'Height', 'Weight', 'Medal'])

# Distribution of ages
plt.figure(figsize=(12, 6))
plt.hist(italy_df_cleaned['Age'], bins=range(10, 70, 1), edgecolor='black')
plt.title('Distribution of Ages of Italian Athletes')
plt.xlabel('Age')
plt.ylabel('Number of Athletes')
plt.grid(axis='y')
plt.show()
```

```python
# Distribution of heights
plt.figure(figsize=(12, 6))
plt.hist(italy_df_cleaned['Height'], bins=range(140, 220, 5), edgecolor='black')
plt.title('Distribution of Heights of Italian Athletes')
plt.xlabel('Height (cm)')
plt.ylabel('Number of Athletes')
plt.grid(axis='y')
plt.show()
```

```python
# Distribution of weights
plt.figure(figsize=(12, 6))
plt.hist(italy_df_cleaned['Weight'], bins=range(40, 140, 5), edgecolor='black')
plt.title('Distribution of Weights of Italian Athletes')
plt.xlabel('Weight (kg)')
plt.ylabel('Number of Athletes')
plt.grid(axis='y')
plt.show()
```

```python
# Medal distribution
medals_counts = italy_df_cleaned['Medal'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(medals_counts, labels=medals_counts.index, autopct='%1.1f%%', startangle=140, colors=['#FFD700', '#C0C0C0', '#CD7F32'])
plt.title('Medal Distribution of Italian Athletes')
plt.show()
```

```python
# Medals by event for Italian athletes
italy_medals_per_event = italy_df_cleaned['Sport'].value_counts()

# Bar chart for medals per event
plt.figure(figsize=(20, 10))
italy_medals_per_event.plot(kind='bar')
plt.title('Number of Medals per Event for Italian Athletes')
plt.xlabel('Event')
plt.ylabel('Number of Medals')
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution of medals (gold, silver, bronze) for each event
italy_medals_distribution = italy_df_cleaned.pivot_table(index='Sport', columns='Medal', aggfunc='size', fill_value=0)

# Stacked bar chart for medal distribution per event
italy_medals_distribution.plot(kind='bar', stacked=True, figsize=(20, 10), color=['#FFD700', '#C0C0C0', '#CD7F32'])
plt.title('Distribution of Medals per Event for Italian Athletes')
plt.xlabel('Event')
plt.ylabel('Number of Medals')
plt.xticks(rotation=90)
plt.legend(title='Medal')
plt.show()
```

```python
winner_Fast = pd.read_csv('./Data/results.csv')
# Grafico 1: Distribuzione delle Medaglie per Anno
plt.figure(figsize=(12, 6))
medals_per_year = winner_Fast['Year'].value_counts().sort_index()
medals_per_year.plot(kind='bar')
plt.title('Distribuzione delle Medaglie per Anno')
plt.xlabel('Anno')
plt.ylabel('Numero di Medaglie')
plt.grid(axis='y')
plt.show()
```

```python
# Grafico 2: Distribuzione delle Medaglie per Nazione
plt.figure(figsize=(20, 6))
medals_per_nation = winner_Fast['Nationality'].value_counts()  # Mostra le prime 20 nazioni
medals_per_nation.plot(kind='bar')
plt.title('Distribuzione delle Medaglie per Nazione')
plt.xlabel('Nazione')
plt.ylabel('Numero di Medaglie')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
# Grafico 3: Distribuzione delle Medaglie per Genere
plt.figure(figsize=(8, 6))
medals_per_gender = winner_Fast['Gender'].value_counts()
medals_per_gender.plot(kind='bar', color=['#66b3ff', '#ff9999'])
plt.title('Distribuzione delle Medaglie per Genere')
plt.xlabel('Genere')
plt.ylabel('Numero di Medaglie')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
# Grafico 4: Distribuzione delle Medaglie per Evento
plt.figure(figsize=(30, 6))
medals_per_event = winner_Fast['Event'].value_counts() 
medals_per_event.plot(kind='bar')
plt.title('Distribuzione delle Medaglie per Evento')
plt.xlabel('Evento')
plt.ylabel('Numero di Medaglie')
plt.xticks(rotation=90)
plt.grid(axis='y')
plt.show()
```

```python
# Grafico 5: Trend delle Medaglie nel Tempo per Nazione
plt.figure(figsize=(14, 8))
top_nations = winner_Fast['Nationality'].value_counts().head(5).index
for nation in top_nations:
    nation_data = winner_Fast[winner_Fast['Nationality'] == nation]
    medals_trend = nation_data['Year'].value_counts().sort_index()
    plt.plot(medals_trend.index, medals_trend.values, marker='o', label=nation)
plt.title('Trend delle Medaglie nel Tempo per Nazione')
plt.xlabel('Anno')
plt.ylabel('Numero di Medaglie')
plt.legend()
plt.grid(axis='y')
plt.show()
```

```python
# Grafico 6: Migliori Atleti (per numero di medaglie)
plt.figure(figsize=(12, 6))
top_athletes = winner_Fast['Name'].value_counts().head(20)  # Mostra i primi 20 atleti
top_athletes.plot(kind='bar')
plt.title('Migliori Atleti per Numero di Medaglie')
plt.xlabel('Atleta')
plt.ylabel('Numero di Medaglie')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
