```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
import plotly.express as px
```

```python
olympics_data = pd.read_csv('./Data/athlete_events.csv');
NOC_data = pd.read_csv('./Data/noc_regions.csv')
# Escludi le competizioni artistiche
olympics_data = olympics_data[olympics_data['Sport'] != 'Art Competitions']

# Ricodifica gli anni dei Giochi Olimpici invernali dopo il 1992
# per abbinarli ai successivi Giochi Olimpici estivi
original = [1994, 1998, 2002, 2006, 2010, 2014]
new = [1996, 2000, 2004, 2008, 2012, 2016]
for orig, new_year in zip(original, new):
    olympics_data['Year'] = olympics_data['Year'].replace(orig, new_year)
```

```python
# Esempio 1: Numero di medaglie per anno
medals_per_year = olympics_data.groupby('Year')['Medal'].count()
plt.figure(figsize=(10, 5))
plt.plot(medals_per_year.index, medals_per_year.values, marker='o')
plt.title('Numero di medaglie per anno')
plt.xlabel('Anno')
plt.ylabel('Numero di medaglie')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.lmplot(x='Height', y='Weight', hue='Sex', data=olympics_data, aspect=3, scatter_kws={'alpha':0.6})
plt.title('Relazione tra altezza e peso degli atleti con linea di regressione')
plt.xlabel('Altezza (cm)')
plt.ylabel('Peso (kg)')
plt.grid(True)
plt.show()
```

```python
# Conta il numero di uomini e donne per anno
counts_sex = olympics_data.groupby(['Year', 'Sex'])['ID'].nunique().unstack(fill_value=0)

# Creazione del grafico a linee
plt.figure(figsize=(12, 6))
# Stili diversi per le linee
sns.lineplot(data=counts_sex['F'], label='Donne', marker='o', linestyle='--', color='blue')
sns.lineplot(data=counts_sex['M'], label='Uomini', marker='s', linestyle='-', color='red')
plt.title('Numero di uomini e donne alle Olimpiadi nel tempo')
plt.xlabel('Anno')
plt.ylabel('Numero di atleti')
plt.grid(True)
plt.show()
```

```python
ages = olympics_data.dropna(subset=['Age'])

# Creazione del grafico boxplot
plt.figure(figsize=(12, 8))

# Definizione della palette di colori per maschi e femmine
colors = {"M": "blue", "F": "red"}

sns.boxplot(x='Year', y='Age', hue='Sex', data=ages, palette=colors)
plt.title('Distribuzione dell\'età degli atleti negli anni')
plt.xlabel('Anno')
plt.ylabel('Età')
plt.grid(True)


# Ruota le etichette dell'asse x verticalmente
plt.xticks(rotation=90)

# Personalizza la legenda
handles, labels = plt.gca().get_legend_handles_labels()
plt.legend(handles, ['Uomini', 'Donne'], title='Sesso')
```

```python
# Step 2: Calcolo del numero di medaglie per ogni ID
medals_count = olympics_data.groupby(['ID', 'Medal']).size().unstack(fill_value=0)

# Step 3: Calcolo del totale medaglie per ogni ID
medals_count['Total'] = medals_count.sum(axis=1)

# Step 4: Selezionare i primi 30 atleti per numero di medaglie totali
top_30_ids = medals_count['Total'].nlargest(30).index

# Step 5: Preparazione dei dati per il grafico
top_30_data = medals_count.loc[top_30_ids, ['Bronze', 'Silver', 'Gold']]
top_30_names = olympics_data[olympics_data['ID'].isin(top_30_ids)][['ID', 'Name']].drop_duplicates().set_index('ID').loc[top_30_ids, 'Name']

colors = ['#CD7F32', '#C0C0C0', '#FFD700']  # Bronze, Silver, Gold

# Step 6: Creazione del grafico a barre impilate
fig, ax = plt.subplots(figsize=(12, 10))
top_30_data.plot(kind='bar', stacked=True, color=colors, ax=ax)

# Impostazioni grafiche
ax.set_title('Medaglie vinte dai primi 30 atleti')
ax.set_xlabel('Atleti')
ax.set_ylabel('Numero di medaglie')
ax.set_xticklabels(top_30_names, rotation=45, ha='right')
ax.legend(['Bronze', 'Silver', 'Gold'], loc='upper right')

plt.tight_layout()
plt.show()
```

```python

ol_medals = olympics_data.dropna(subset=['Medal']).drop_duplicates()
medal_counts = ol_medals.groupby('NOC').size().reset_index(name='Medal Count')

# Filtrare le medaglie del Brasile
brazil_medals = ol_medals[ol_medals['NOC'] == 'BRA']

# Stampa delle medaglie del Brasile
print("Medaglie del Brasile:")
for index, row in brazil_medals.iterrows():
    print(f"Medaglia: {row['Medal']} - Anno: {row['Year']}- Atleta: {row['Name']}")

print(medal_counts.to_string())

print("\nNumero di duplicati dopo la join:", medal_counts.duplicated().sum())

# Unisci i dati delle medaglie con il file di mapping NOC-region
medal_counts = medal_counts.merge(NOC_data, on='NOC', how='left')

# Crea la mappa
fig = px.choropleth(
    medal_counts,
    locations='region',
    locationmode='country names',
    color='Medal Count',
    hover_name='region',
    color_continuous_scale='Viridis',
    title='Total Olympic Medals by Country (1920-2016)'
)

# Mostra la mappa
fig.show()
```
